```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
target_names = iris.target_names
fnames = iris.feature_names
```

```python
plt.xlabel(fnames[2])
plt.ylabel(fnames[3])
plt.scatter(iris.data[:,2][:50], iris.data[:,3][:50],c='blue',marker='*',label=target_names[0])
plt.scatter(iris.data[:,2][50:100],iris.data[:,3][50:100],c='green',marker='+',label=target_names[1])
plt.scatter(iris.data[:,2][100:150],iris.data[:,3][100:150],c='black',marker='.',label=target_names[2])
plt.legend()
```
